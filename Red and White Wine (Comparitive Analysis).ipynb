```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
red_wine_df = pd.read_csv('winequality-red.csv', sep = ';')
white_wine_df = pd.read_csv('winequality-white.csv', sep = ';')
```

```python
red_wine_df
```

```python
white_wine_df
```

```python
sns.set(rc = {'figure.figsize' : (20,18)})
sns.heatmap(red_wine_df.corr(), fmt = '.2f', annot = True)
```

```python
sns.set(rc = {'figure.figsize' : (20,18)})
sns.heatmap(white_wine_df.corr(), fmt = '.2f', annot = True)
```

```python
red_wine_df['type'] = 'Red'
white_wine_df['type'] = 'White'
```

```python
wines_df = pd.concat([red_wine_df,white_wine_df])
wines_df
```

```python
sns.set_style('whitegrid')
sns.countplot(data = wines_df, x = 'quality', hue = 'type')
```

```python
sns.set(rc = {'figure.figsize' : (15,12)})
g = sns.kdeplot(red_wine_df.quality, shade = True, color = 'r', label = 'Red Wine')
g = sns.kdeplot(white_wine_df.quality, shade = True, color = 'b', label = 'White Wine')
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'alcohol', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'density', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'residual sugar', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```

```python
sns.regplot(x = 'alcohol', y = 'residual sugar', data = red_wine_df)
```

```python
sns.regplot(x = 'alcohol', y = 'residual sugar', data = white_wine_df)
```

```python
sns.set(rc = {'figure.figsize' : (15,12)})
g = sns.kdeplot(red_wine_df['residual sugar'], shade = True, color = 'r', label = 'Red Wine')
g = sns.kdeplot(white_wine_df['residual sugar'], shade = True, color = 'b', label = 'White Wine')
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'total sulfur dioxide', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'free sulfur dioxide', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```

```python
sns.boxplot(data = wines_df, x='quality', y = 'citric acid', hue  = 'type', palette = ['r', 'w'], showfliers = False)
```
