# Frequency Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

### Lets create a random normal distribution to demonstrate basic statistic visualization

```python
x = np.random.normal(50, 5, 3000)
```

#### first args is MEAN
#### second args is SD
#### third args is SIZE

```python
x
```

```python
max(x)
```

```python
min(x)
```

```python
max(x)-min(x)
```

#### Lets Plot a Histogram to plot this normal distribution

```python
sns.distplot(x, kde = False, bins = 50, hist_kws = dict(edgecolor = 'k'))
plt.xlabel('Intervals')
plt.ylabel('Frequency')
plt.show()
```

### Relative Frequency Histogram Plot
#### A relative frequency histogram is a type of histogram that displays the proportion of observations that fall within each bin.

```python
sns.distplot(x, kde = False, bins = 50, hist_kws = dict(edgecolor = 'k'), norm_hist = True)
plt.xlabel('Intervals')
plt.ylabel('Relative Frequency')
plt.show()
```

# Density Plot

```python
sns.distplot(x, kde = True, bins = 30, hist_kws = dict(edgecolor = 'k'), norm_hist = True)
plt.ylabel('Density');
```

```python
sns.distplot(x, kde = True, bins = 30, hist_kws = dict(edgecolor = 'k'), hist = False)
plt.ylabel('Density')
plt.axvline(x.mean(), color = 'b', linestyle = '--')
plt.axvline(np.median(x), color = 'r', linestyle = ':');
```

#### The pink dotted line in the center represents the median
#### The blue dashed line in the center represents the median

# Box Plot 

```python
myList = np.asarray([1,2,2,2,4,5,6,6,6,9,10,10,10,11,14,14,15,18,19,20])
np.percentile(myList, [25,50,75])
```

```python
df = pd.DataFrame(myList, columns = ['Sample Data'])
sns.set_style('whitegrid')
sns.boxplot(data = df, )
plt.ylabel('Data Points')
```

#### A boxplot is a graphical representation of a dataset that displays the distribution ofm the data based on its quartiles. It shows the median (50th percentile), the upper and lower quartiles (25th and 75th percentiles), and any outliers that may be present. Boxplots are useful for identifying the spread and central tendency of a dataset and for detecting outliers.
